```python
import numpy as np
def add_bias_variable(x):
    n=x.shape[0]
    x_b = np.concatenate((np.ones((n,1)),x),axis=1)
    return x_b
def prediction(x,theta):
    predictions=np.dot(x,theta)
    return predictions
def error(predictions,y):
    e=0
    for i in range(len(predictions)):
        e+=(predictions[i]-y[i])**2
    return e/(2*len(predictions))
def gradient(x,prediction,y):
    len = x.shape[1]
    k=x.shape[0]
    temp =[]
    for j in range(len):
        theta_j=0
        for i in range(k):
            theta_j+=(prediction[i]-y[i])*x[i][j]
        temp.append(theta_j/k)
    return np.array(temp)
def gradient_descent(X,Y, theta, learning_rate, iterations):
        """Performs gradient descent to optimise theta. """
        for i in range(iterations):
            predictions = prediction(X,theta)
            grad = gradient(X, predictions, Y)
            theta -= learning_rate* grad
            if i% 100 == 0:
                print(f"Iteration {i}: Error ={error(predictions, Y)}")
        return theta
x = 2*np.random.rand(100,1)
y = 4 + 3*x + np.random.randn(100,1)
print(x.shape)
print(y.shape)
learning_rate = 0.1
iterations = 1000
x = add_bias_variable(x)
theta = np.zeros((x.shape[1],1))
print(theta.shape)
theta = gradient_descent(x, y, theta, learning_rate, iterations)
print("Optimised Theta:", theta)
```

```python
def logistic_regression(x,y,theta,learning_rate=0.1,iterations=1000):
  errors=[]
  for i in range(iterations):
    y_pred = 1/(1+np.exp(-np.dot(x,theta)))
    error = -sum(y*np.log(y_pred)+(1-y)*np.log(1-y_pred))/len(y)
    errors.append(error)
    gradient = np.dot(x.T,(y_pred-y))/len(y)
    theta -= learning_rate*gradient
  return theta,errors
```

```python
def TP(y,y_pred):
    return sum((y==1)&(y_pred==1))
def TN(y,y_pred):
    return sum((y==0)&(y_pred==0))
def FP(y,y_pred):
    return sum((y==0)&(y_pred==1))
def FN(y,y_pred):
    return sum((y==1)&(y_pred==0))
def accuracy(y,y_pred):
    return (TP(y,y_pred)+TN(y,y_pred))/len(y)
def precision(y,y_pred):
    return (TP(y,y_pred)/(TP(y,y_pred)+FP(y,y_pred)))
def recall(y,y_pred):
    return TP(y,y_pred)/(TP(y,y_pred)+FN(y,y_pred))
def f1_score(y,y_pred):
    return 2*precision(y,y_pred)*recall(y,y_pred)/(precision(y,y_pred)+recall(y,y_pred))
                                                              
```
